```python
from pathlib import Path
import pandas as pd
import requests


def get_url(url):
    """Return the final URL after following redirects."""
    try:
        r = requests.get(url, allow_redirects=False)
        return r.headers["Location"]
    except:
        return url


# Function to return the url minus the protocol
def get_domain(url):
    """Return the domain of a URL."""
    return url.split("//")[-1].split("/")[0]


competitors = []
us = set()
for file in Path("../data/true_competitors").glob("*.csv"):
    filename = file.name.split("-")[-1][:-4]
    site = f"https://{filename}"
    site = get_url(site)
    site = get_domain(site)
    us.add(site)

    df = pd.read_csv(file)
    if site not in df["Top Competitor URLs"].values:
        print("Site not listed as competitor:", site)
    competitors.append(df["Top Competitor URLs"])

competitors = set.union(*map(set, competitors))
```

```python
for u in us:
    if u not in competitors:
        print(f"Site not listed as competitor: {u}")
        competitors.add(u)
    
```
